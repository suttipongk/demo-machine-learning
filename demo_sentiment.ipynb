```
import numpy as np 
import pandas as pd 
import re
import nltk 
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Download the data

```
data_source_url = "https://raw.githubusercontent.com/kolaveridi/kaggle-Twitter-US-Airline-Sentiment-/master/Tweets.csv"
airline_tweets = pd.read_csv(data_source_url)
```

```
airline_tweets.head()
```

# 2. extract Text and Label from main table

```
texts = airline_tweets.iloc[:, 10].values
labels = airline_tweets.iloc[:, 1].values
```

# 3. Clean text


```
processed_features = []
for sentence in texts:
    # Remove all the special characters
    processed_feature = re.sub(r'\W', ' ', str(sentence))
    processed_feature = processed_feature.lower()
    processed_features.append(processed_feature)
```

# 4. Find (features) and extract data 2 parts


```
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer (max_features=2500, min_df=7, max_df=0.8)
processed_features = vectorizer.fit_transform(processed_features)

X_train, X_test, y_train, y_test = train_test_split(processed_features, labels, test_size=0.2, random_state=0)
```

# 5. Train model with text and label from Training set


```
from sklearn.ensemble import RandomForestClassifier

text_classifier = RandomForestClassifier(n_estimators=200, random_state=0)
text_classifier.fit(X_train, y_train)
```

# Testing Model

```
from sklearn.metrics import accuracy_score
predictions = text_classifier.predict(X_test)
print(accuracy_score(y_test, predictions))
```

# deploy

```
test_feature = vectorizer.transform(['My flight this morning is a total disaster.'])
text_classifier.predict(test_feature)
```
